# $F_s(k,t)$

## True Trajectory

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize = (8, 6), dpi = 300)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.tick_params(direction = 'in', width = 3, length = 5, pad = 10, labelsize = 15)

ROOT = "/mnt/CALC_SSD/bgkang/inherent/trajectories/NVT"
timestep = 1.0E-1

for idx, temp in enumerate(["300K", "400K", "500K"]):
    data = np.stack([np.genfromtxt(f"{ROOT}/{temp}/ensemble{e}/true_fskt.out") for e in ["00", "01", "02", "03"]])
    avg = data.mean(axis = 0)
    std = data.std(axis = 0)
    
    time = timestep * avg[:,0]
    
    ax.plot(time, avg[:,5], linewidth = 3, label = temp, marker = 'o', markerfacecolor = '#FFFFFF', markersize = 10, markeredgewidth = 3, color = f"C{idx}")
    ax.fill_between(time, avg[:,5] + std[:,5], avg[:,5] - std[:,5], alpha = 0.3, color = f"C{idx}")

ax.set_xlabel("Time (ps)", fontsize = 20)
ax.set_ylabel("$F_s(k,t)$", fontsize = 20)

ax.set_ylim((-0.1, 1.1))
ax.set_xscale('log')

ax.legend(fontsize = 20, frameon = False)

plt.tight_layout()
plt.savefig('./true_fskt.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```

## Combined

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize = (8, 6), dpi = 300)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.tick_params(direction = 'in', width = 3, length = 5, pad = 10, labelsize = 15)

ROOT = "/mnt/CALC_SSD/bgkang/inherent/trajectories/NVT"
timestep = 1.0E-1

for idx, temp in enumerate(["300K", "400K", "500K"]):
    true = np.stack([np.genfromtxt(f"{ROOT}/{temp}/ensemble{e}/true_fskt.out") for e in ["00", "01", "02", "03"]])
    true_avg = true.mean(axis = 0)
    true_std = true.std(axis = 0)
    inherent = np.stack([np.genfromtxt(f"{ROOT}/{temp}/ensemble{e}/inherent_fskt.out") for e in ["00", "01", "02", "03"]])
    inherent_avg = inherent.mean(axis = 0)
    inherent_std = inherent.std(axis = 0)

    time = timestep * true_avg[:,0]
    
    ax.plot(time, true_avg[:,5], linewidth = 3, label = temp + r"$_{\rm true}$", marker = 'o', markerfacecolor = '#FFFFFF', markersize = 10, markeredgewidth = 3, color = f"C{idx}")
    ax.fill_between(time, true_avg[:,5] + true_std[:,5], true_avg[:,5] - true_std[:,5], alpha = 0.3, color = f"C{idx}")
    ax.plot(time, inherent_avg[:,5], linewidth = 3, label = temp + r"$_{\rm inherent}$", marker = '^', markerfacecolor = '#FFFFFF', markersize = 10, markeredgewidth = 3, color = f"C{idx}")
    ax.fill_between(time, inherent_avg[:,5] + inherent_std[:,5], inherent_avg[:,5] - inherent_std[:,5], alpha = 0.3, color = f"C{idx}")

ax.set_xlabel("Time (ps)", fontsize = 20)
ax.set_ylabel("$F_s(k,t)$", fontsize = 20)

ax.set_ylim((-0.1, 1.1))
ax.set_xscale('log')

ax.legend(fontsize = 20, frameon = False)

plt.tight_layout()
plt.savefig('./combined_fskt.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```

# Energy Profile

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize = (8, 6), dpi = 300, width_ratios = (0.8, 0.2))

ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].spines['left'].set_linewidth(3)
ax[0].spines['bottom'].set_linewidth(3)
ax[0].tick_params(direction = 'in', width = 3, length = 5, pad = 10, labelsize = 15)
ax[0].set_xlabel("Time (ps)", fontsize = 20)
ax[0].set_ylabel(r"$U(t)_{\rm inherent}$ (eV)", fontsize = 20)

ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['left'].set_linewidth(3)
ax[1].spines['bottom'].set_visible(False)
ax[1].tick_params(direction = 'in', width = 3, length = 5, pad = 10, labelsize = 15)
ax[1].set_yticks([])
ax[1].set_xticks([])

ROOT = "/mnt/CALC_SSD/bgkang/inherent/trajectories/NVE"
timestep = 1.0E0

temp = "300K"

for e in ["00", "01", "02", "03"]:
    data = np.genfromtxt(f"{ROOT}/{temp}/ensemble{e}/energy_profile.out")

    time = timestep * data[:,0]

    ax[0].plot(time, data[:,1], linewidth = 3, label = f"ensemble{e}", alpha = 0.7)


    ax[1].hist(data[:,1], bins = 100, orientation = 'horizontal', alpha = 0.7)

ax[0].legend(fontsize = 20, frameon = False)

plt.tight_layout()
plt.savefig('./energy_profile_300K.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```

```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize = (8, 6), dpi = 300)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.tick_params(direction = 'in', width = 3, length = 5, pad = 10, labelsize = 15)
ax.set_xlabel(r"$U(t)_{\rm inherent}$ (eV)", fontsize = 20)
ax.set_ylabel(r"$P(U(t)_{\rm inherent})$", fontsize = 20)

ROOT = "/mnt/CALC_SSD/bgkang/inherent/trajectories/NVE"

temp = "300K"
x = np.linspace(-6015, -5996, 200)

for idx, e in enumerate(["00", "01", "02", "03"]):
    data = np.genfromtxt(f"{ROOT}/{temp}/ensemble{e}/energy_profile.out")

    ax.hist(data[:,1], density = True, bins = 100, alpha = 0.3, label = f"ensemble{e}", color = f"C{idx}")

    kde = gaussian_kde(data[:,1])
    ax.plot(x, kde(x), linewidth = 3, linestyle = '--', color = f"C{idx}")

ax.set_xticklabels(range(-6015, -5996, 2), rotation = 45)

ax.legend(fontsize = 20, frameon = False)

plt.tight_layout()
# plt.savefig('./energy_kde_300K.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```
